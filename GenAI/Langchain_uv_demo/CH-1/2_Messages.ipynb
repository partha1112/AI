```python
import os
from dotenv import load_dotenv
load_dotenv()
if os.environ.get("OPENAI_API_KEY"):
    print("Open AI API Key Available")
else:
    raise ValueError("Open AI API Key No available")
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_openai import ChatOpenAI

my_messages=[
    SystemMessage("You are a gen-z assistance who always answer in fun way "),
    HumanMessage(content="Bro, tell me some fun factor about NASA")
]

## my_messages format only support OpenAI, Each LLM may have different format 

llm_openai = ChatOpenAI(model="gpt-5-mini", temperature=0)

llm_openai.invoke(my_messages).content
```

```python
##Prompts - Dynamic way of sending messages which support all kind of LLMs
from langchain_core.prompts import PromptTemplate

user_input = input("Enter your topic")

dynamic_promopt = PromptTemplate.from_template("Write a fun fact about {topic}")
ready_prompt = dynamic_promopt.invoke({"topic" : user_input})

llm_openai.invoke(ready_prompt).content
```

```python
## If you want to set the tone then use ChatPromptTemplate

from langchain_core.prompts import  ChatPromptTemplate

promt_template = ChatPromptTemplate.from_messages([
    ("system","You are a funny gen-z, answer infun way"),
    ("user","Write a fun fact about {topic}")
])

user_input2=input("Enter topic ")

ready_prompt2=promt_template.invoke({"topic",user_input2})
llm_openai.invoke(ready_prompt2)
```
