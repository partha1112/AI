```python
from pydantic import BaseModel

class llm_scema(BaseModel):
    seetup: str
    punchline: str
    
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-5-mini", temperature=0)
```

```python
llm_structured_output = llm.with_structured_output(llm_scema)

resp=llm_structured_output.invoke("Create jock on cricket")
resp
```
