# **First Chain**

```python
# Task 1 : Prompt
from langchain_core.prompts import ChatPromptTemplate

promt_template = ChatPromptTemplate.from_messages([
    ("system","you are a helpful assistant"),
    ("human","{input}")
])
```

```python
# Task 2 : LLM
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model='gpt-5-mini', temperature=0)
```

```python
# Task 3 : output parser

from langchain_core.output_parsers import StrOutputParser
str_parser = StrOutputParser()

```

# **Manual Invocation**

```python
template = promt_template.invoke({"input","Tell me some intresting thing about Avengers"})

result = llm.invoke(template)

result.content
```

## **Using Chain**

```python
chain = promt_template | llm | str_parser
chain.invoke({"input","Tell me some intresting thing about Avengers"})
```

```python
#Using Runnable  Sequence

from langchain_core.runnables import RunnableSequence

chain2=RunnableSequence(promt_template,llm,str_parser)
chain2.invoke("Tell me some intresting thing about Avengers")
```
