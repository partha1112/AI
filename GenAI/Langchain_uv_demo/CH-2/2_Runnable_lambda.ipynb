# **Runnable Lamba**

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()
```

```python
promt_template = ChatPromptTemplate.from_messages([
    ("system","you are a helpful assistant"),
    ("human","Write some funy thing about  {name}")
])

llm = ChatOpenAI(model="gpt-5-mini",temperature=0)


str_parser= StrOutputParser()
```

```python
from langchain_core.runnables import RunnableLambda

def dic_maker(text:str) ->  dict:
    return {"text": text}

dic_maker_runnable = RunnableLambda(dic_maker)
```

```python
post_prompt = ChatPromptTemplate([
    ("system", "you are a social media poist generator"),
    ("human","Create a post for the following text for Instagram : {text}")
])
```

## **Generate Post using Runnable**

```python
chain = promt_template | llm | str_parser | dic_maker_runnable | post_prompt | llm | str_parser
```

```python
response = chain.invoke({"name": "Virat Kohli"})
```
