# **Parallel Chain**

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()
```

```python
#Task 1

promt_template = ChatPromptTemplate.from_messages([
    ("system","you are a movie reviewer"),
    ("human","Write some review on the movie : {name}")
])

llm = ChatOpenAI(model="gpt-5-mini",temperature=0)


str_parser= StrOutputParser()
```

## Parallel Chain 1

```python
from langchain_core.runnables import RunnableLambda, RunnableParallel

def dic_maker(text:str) ->  dict:
    return {"text": text}

dic_maker_runnable = RunnableLambda(dic_maker)
```

```python
facebbok_prompt = ChatPromptTemplate.from_messages([
    ("system","you are a funny writter who will create funny post for Facebook"),
    ("human","create a facebook post for the following text : {text}")

])

chain_facebook = facebbok_prompt | llm | str_parser
```

## Parallel Chain 2

```python
def x_chain(text:dict):

    text=text["text"]

    x_prompt = ChatPromptTemplate.from_messages([
    ("system","you are a serious content writter on x platform"),
    ("human","create a serious post for the following text : {text}")

    ])

    chain_x = x_prompt| llm | str_parser

    result = chain_x.invoke(text)

    return result

x_chain_runnable = RunnableLambda(x_chain)
```

## Final Parallel Chain Orchestration

```python
final_chain = (
    promt_template      |
    llm                 |
    str_parser          |
    dic_maker_runnable  |
    RunnableParallel( branches = {"facebook" : chain_facebook, "X_platform" : x_chain_runnable})
)
```

```python
final_response = final_chain.invoke("Padaiyappa")
```

# Beautify Final response

```python
def beautify(final_response:dict)->dict:
    facebook = final_response['branches']['facebook']
    X_platform = final_response['branches']['X_platform']
    return {"facebook":facebook, "X_platform":X_platform}

beauify_runnable = RunnableLambda(beautify)
```

```python
beatifyl_chain = final_chain | beauify_runnable
```
